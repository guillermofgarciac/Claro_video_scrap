```python

wd = webdriver.Chrome('./chromedriver')#Driver de crome
wd.get(url)# Aca utilizamos un get para el driver
wd.maximize_window()#maximizamos la ventana
slider_xpath = '//*[@id="app"]/div/div[1]/div[2]/div/div/section[1]/div[2]/div'#esta es direccion del slider de generos
slider = WebDriverWait(wd, 10).until(EC.presence_of_element_located((By.XPATH, slider_xpath)))#localizamos el elemento
lis = slider.find_elements_by_tag_name('li')#guardamos la etiqueta
# El primer elemento remite al catálogo completo
lis.pop(0)#elminamos el primero, porque no nos lleva a nuestro lugar deseado
ids = list()# lista de IDS
```

```python
for i in range(len(lis)):
    try:
        lis[i].click()# intentar hacer click en nuestra licalizacion
        time.sleep(5)#lo hago esperar 5 segundos porque mi maquina no lo corria correctamente, pero este tiempo se puede modificar
    except:
        result = False #Si no funciona pasa esto
        while result == False:#mientras pase esto
            try:
                wd.find_element_by_class_name('scrollinghotspotright').click()#nuestro driver va a scrolear generos
                time.sleep(5)#otra vez tiempor porque a veces no llega a cargar lo suficiente
                lis[i].click()#clickeamos
                result = True
            except:
                pass
    plays = WebDriverWait(wd, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'play')))#busca los plays de cada item encontrado en el url
    for i in range(len(plays)):
        plays[i].click()#clickea en cada uno
        if re.findall(r'/(\d+)$', wd.current_url)[0] not in ids:#con un regex sacamos el id de la pelicula
                print(wd.current_url)
                ids.append(re.findall(r'/(\d+)$', wd.current_url)[0])
                print(ids[-1])
        wd.back()#volvemos atras
        plays = WebDriverWait(wd, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'play')))
    wd.back()
    elems = WebDriverWait(wd, 10).until(EC.presence_of_element_located((By.XPATH, slider_xpath)))#volvemos a scrolear nuestro carril de generos
    lis = elems.find_elements_by_tag_name('li')
    lis.pop(0)
```

```python
lista_urls = []
def id_api(lista):
    url = 'https://mfwkweb-api.clarovideo.net/services/content/data?device_id=web&device_category=web&device_model=web&device_type=web&device_so=Edge&format=json&device_manufacturer=generic&authpn=webclient&authpt=tfg1h3j4k6fd7&api_version=v5.92&region=argentina&HKS=9flfjj6v3gegp54nn3cde1l030&group_id='
    for item in lista:#por cada item en la lista a la que le apliquemos la funcion
        item = url + item #agrega el ID de cada pelicula en la URL de la api
        lista_urls.append(item)
    return lista_urls# devuelve una lista de URLs
```

```python
json_list = []
def scraping(lista_urls):
    for url in lista_urls:#por cada item en esa lista
        r = requests.get(url)#utilizamos requests para solicitar acceso
        data = r.text #lo pasamos a texto
        json_data = json.loads(data)#lo pasamos a json
        json_list.append(json_data)y lo pendeamos a una lista
        
    return json_list
```

```python
peliculas = []
def json_information(lista_json):#con esta funcion limpiamos los json para que quede nuestra info legible
    for item in lista_json:
        item_1 = item['response']['group']['common']
        item_limpio = dict(itertools.islice(item_1.items(), 7))
        peliculas.append(item_limpio)
    return peliculas
```
